## Step 2: Understanding NumPy Data Types

Setting dtype manually is a **best practice** because it:
- Improves performance
- Makes programs more memory efficient
- Prevents unexpected type conversions

### Available Data Types:

| Category | Data Types |
|----------|-----------|
| Integer | int8, int16, int32, int64 |
| Float | float16, float32, float64 |
| String | str_, U# (unicode string) |
| Boolean | bool_ |
| Object | object_ (mixed types) |

Let's see how to use each data type:

```python
# Create an int8 array
arr = np.array([2, 4, 6, 8], dtype=np.int8)
print("Data type:", arr.dtype)
print("Array:", arr)
print(f"Memory usage: {arr.nbytes} bytes")
print()
```

### Integer Types - int8 (1 byte per element)
The int8 type can store integers from -128 to 127.

```python
# Create an int16 array (note: values beyond int8 range can overflow)
arr = np.array([125, 126, 127, 128], dtype=np.int16)
print("Data type:", arr.dtype)
print("Array:", arr)
print(f"Memory usage: {arr.nbytes} bytes")
print()
```

### Integer Types - int16 (2 bytes per element)
The int16 type can store integers from -32768 to 32767, using twice the memory of int8.

```python
# Create a float16 array
arr = np.array([2, 4, 6, 8], dtype=np.float16)
print("Data type:", arr.dtype)
print("Array:", arr)
print(f"Memory usage: {arr.nbytes} bytes")
print()
```

### Float Types - float16 (2 bytes per element)
Floating-point numbers with reduced precision to save memory. Note the precision loss compared to float64!

```python
# Create a boolean array (non-zero = True, zero = False)
arr = np.array([2, 4, 6, 8], dtype=np.bool_)
print("Data type:", arr.dtype)
print("Array:", arr)
print(f"Memory usage: {arr.nbytes} bytes")
print()
```

### Boolean Type - bool_ (1 byte per element)
Represents True/False values. Non-zero numbers are converted to True, zero to False.

```python
# Create a string array
arr = np.array(['Hello', 'python'], dtype=np.str_)
print("Data type:", arr.dtype)
print("Array:", arr)
print(f"Memory usage: {arr.nbytes} bytes")
print()
```

### String Type - str_ (unicode strings)
Stores text strings. Each character takes multiple bytes depending on the character set.

```python
# Create an object array with mixed types
arr = np.array(['Hello', True, 2, 4.4], dtype=np.object_)
print("Data type:", arr.dtype)
print("Array:", arr)
print(f"Memory usage: {arr.nbytes} bytes")
print()
```

### Object Type - object_ (mixed types)
Can store mixed data types (integers, floats, strings, booleans, etc.) in a single array. Uses more memory.

```python
# Convert array data types using astype()
arr = np.array([2.2, 4.4, 6.6, 8.1])  # Original float64 array
print("Original array:")
print(f"Data type: {arr.dtype}")
print(f"Array: {arr}")
print(f"Memory usage: {arr.nbytes} bytes\n")

# Convert to int16 (note: decimal values are truncated)
newArr = arr.astype(np.int16)
print("Converted to int16:")
print(f"Data type: {newArr.dtype}")
print(f"Array: {newArr}")
print(f"Memory usage: {newArr.nbytes} bytes")
```

## Summary & Key Takeaways

### Memory Usage Comparison
- **int8**: 1 byte per element (range: -128 to 127)
- **int16**: 2 bytes per element
- **int32**: 4 bytes per element
- **int64**: 8 bytes per element
- **float16**: 2 bytes per element
- **float32**: 4 bytes per element
- **float64**: 8 bytes per element
- **bool_**: 1 byte per element
- **str_**: Varies based on string length
- **object_**: Stores references (uses more memory)

### Best Practices

1. **Choose appropriate data types** - Use int8 instead of int64 if you only need small numbers
2. **Avoid object_ type** - Use specific types for better performance
3. **Check memory usage** - Use `.nbytes` to monitor memory consumption
4. **Consider precision** - float16 uses less memory but has lower precision than float64
5. **Convert when needed** - Use `.astype()` to change types after creation

### Performance Impact
- Using smaller data types (int8 vs int64) can reduce memory by 8x
- Operations on smaller types are faster
- Important for large datasets and real-time processing
