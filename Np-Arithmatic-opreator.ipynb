```python
import numpy as np
```

```python
#scalar arithmetic
arr = np.array([2,4,6])
print(arr * 2)
print(arr + 2)
print(arr - 2)
print(arr * 2)
print(arr / 2)
print(arr ** 2)
```

```python
#Vectorized math function 
arr = np.array([2,4,6])
print(np.sqrt(arr))
```

```python
arr = np.array([2.6 ,4.7 ,6.3])
print(np.round(arr))
```

```python
arr = np.array([2.6 ,4.7 ,6.3])
print(np.floor(arr))
```

```python
arr = np.array([2.6 ,4.7 ,6.3])
print(np.ceil(arr))
```

```python
#Element -wise arithmatic
arr1 = np.array([2, 4, 6])
arr2 = np.array([1, 2, 3])

print(arr1 + arr2)
print(arr1 - arr2)
print(arr1 * arr2)
```

```python
#Comparison operator
Marks = np.array([45, 74, 66, 87])
print(Marks >=50)
print(Marks < 50)

Marks[0] = 30
print(Marks)

Marks[Marks < 50] = 0
print(Marks)
```
