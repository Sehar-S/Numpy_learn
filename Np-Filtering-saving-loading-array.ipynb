```python
import numpy as np
```

```python
#filtering : process of selecting elements from an array that match a given condition
marks = np.array([ [44,56,92,75,32,89 ],
                   [83,58,39,68,79,95 ] ])

passStudents = marks[marks > 50]
#print(passStudents)
failStudents = marks[marks < 50]
#print(failStudents)
cgrad = marks[(marks >= 50) & (marks <= 70)]
print(cgrad)
bgrad = marks[(marks >= 70) & (marks <= 80)]
print(bgrad)
agrad = marks[(marks >= 80) & (marks <= 100)]
print(agrad)
```

```python
#saving and loading arrays

marks = np.array([ [44,56,92,75,32,89 ],
                   [83,58,39,68,79,95 ] ])


marks2 = np.array([ [26,56,92,75,32,89 ],
                   [27,58,39,68,79,95 ] ])

#np.save('marks',marks) # for saving single array 
np.savez('MultipleMarks',marks,marks2) # for saving multiple array
print('saving marks')
```

```python
#Compressed file save for less memory 
np.savez_compressed('MultipleMarks-compressed',marks,marks2) # for saving multiple array
```

```python
#load compressed file
array = np.load('MultipleMarks-compressed.npz')
print(array)
```

```python
#access compressed file data using keys 
newmarks = array['arr_0']
#print(newmarks)

newmarks2 = array['arr_1']
print(newmarks2)
```
